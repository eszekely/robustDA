```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

%matplotlib inline
```

### 1. Radiative forcing

Radiative forcing from solar, volcanic, anthropogenic and all combined for RCP4.5 and RCP8.5.

```python
startDate, endDate = 1870, 2100

dirFiles = "./../data/mount/_DATA/reg_glob/FORCING_CMIP5/global/ALLDATA_30May2010/"

# Read RCP4.5
filename = dirFiles + 'RCP45_MIDYEAR_RADFORCING.xls'
df_rcp45_radforcing = pd.read_excel(filename, sheet_name = 3, skiprows = 59, header = 0, index_col = 0)
df_rcp45_radforcing.index.name = 'YEAR'
df_rcp45_radforcing = df_rcp45_radforcing.loc[startDate : endDate]
df_rcp45_radforcing_all = df_rcp45_radforcing['TOTAL_INCLVOLCANIC_RF']
df_rcp45_radforcing_anthro = df_rcp45_radforcing['TOTAL_ANTHRO_RF']
df_rcp45_radforcing_volcanic = df_rcp45_radforcing['VOLCANIC_ANNUAL_RF']
df_rcp45_radforcing_solar = df_rcp45_radforcing['SOLAR_RF']
df_rcp45_radforcing_GHG = df_rcp45_radforcing['GHG_RF']
df_rcp45_radforcing_CO2 = df_rcp45_radforcing['CO2_RF']
df_rcp45_radforcing_CH4 = df_rcp45_radforcing['CH4_RF']

# Read RCP8.5
filename = dirFiles + 'RCP85_MIDYEAR_RADFORCING.xls'
df_rcp85_radforcing = pd.read_excel(filename, sheet_name = 3, skiprows = 59, header = 0, index_col = 0)
df_rcp85_radforcing.index.name = 'YEAR'
df_rcp85_radforcing = df_rcp85_radforcing.loc[startDate : endDate]
df_rcp85_radforcing_all = df_rcp85_radforcing['TOTAL_INCLVOLCANIC_RF']
df_rcp85_radforcing_anthro = df_rcp85_radforcing['TOTAL_ANTHRO_RF']
df_rcp85_radforcing_volcanic = df_rcp85_radforcing['VOLCANIC_ANNUAL_RF']
df_rcp85_radforcing_solar = df_rcp85_radforcing['SOLAR_RF']
df_rcp85_radforcing_GHG = df_rcp85_radforcing['GHG_RF']
df_rcp85_radforcing_CO2 = df_rcp85_radforcing['CO2_RF']
df_rcp85_radforcing_CH4 = df_rcp85_radforcing['CH4_RF']

fig, (ax1, ax2) = plt.subplots(1,2, figsize=(16, 4))
end = 2100
df_rcp85_radforcing_subset = df_rcp85_radforcing.loc[:end]

ax1.plot(df_rcp45_radforcing_anthro, label = 'anthropogenic forcing')
ax1.plot(df_rcp45_radforcing_volcanic, label = 'volcanic forcing')
ax1.plot(df_rcp45_radforcing_solar, label = 'solar forcing')
ax1.plot(df_rcp45_radforcing_volcanic + df_rcp45_radforcing_solar, label = 'natural forcing (volcanic + solar)')
ax1.plot(df_rcp45_radforcing_all, 'k', label = 'all forcings')
ax1.legend()
ax1.set_ylabel("Radiative forcing")
ax1.set_title("RCP 4.5")

ax2.plot(df_rcp85_radforcing_anthro, label = 'anthropogenic forcing')
ax2.plot(df_rcp85_radforcing_volcanic, label = 'volcanic forcing')
ax2.plot(df_rcp85_radforcing_solar, label = 'solar forcing')
ax2.plot(df_rcp85_radforcing_volcanic + df_rcp85_radforcing_solar, label = 'natural forcing (volcanic + solar)')
ax2.plot(df_rcp85_radforcing_all, 'k', label = 'all forcings')
ax2.legend()
ax2.set_ylabel("Radiative forcing")
ax2.set_title("RCP 8.5");
```

#### 1.1. Plot the different radiative forcings

```python
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(18,4), gridspec_kw = {'width_ratios':[3, 1]})
ax1.plot(np.array(df_rcp45_radforcing_volcanic), 'k', marker = '.', markeredgecolor = 'b')
ax1.set_title("Volcanic radiative forcing")
ax1.set_ylabel("Volcanic forcing")
ax1.set_xlabel("Year")
ax1.xaxis.set_ticks([0, 30, 80, 135, 230])
ax1.xaxis.set_ticklabels([1870, 1900, 1950, 2005, 2100])
ax2.plot(np.array(df_rcp45_radforcing_volcanic), df_rcp45_radforcing_volcanic.index, 'b.')
ax2.set_xlabel("Volcanic forcing")
ax2.set_ylabel("Year");
```

```python
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(18,4), gridspec_kw = {'width_ratios':[3, 1]})
ax1.plot(np.array(df_rcp45_radforcing_solar), 'k', marker = '.', markeredgecolor = 'b')
ax1.set_title("Solar radiative forcing")
ax1.set_ylabel("Solar forcing")
ax1.set_xlabel("Year")
ax1.xaxis.set_ticks([0, 30, 80, 135, 230])
ax1.xaxis.set_ticklabels([1870, 1900, 1950, 2005, 2100])
ax2.plot(np.array(df_rcp45_radforcing_solar), df_rcp45_radforcing_solar.index, 'b.')
ax2.set_xlabel("Solar forcing")
ax2.set_ylabel("Year");
```

```python
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(18, 4), gridspec_kw = {'width_ratios':[3, 1]})
ax1.plot(np.array(df_rcp45_radforcing_anthro), 'k', marker = '.', markeredgecolor = 'b', label = "RCP 4.5")
ax1.plot(np.array(df_rcp85_radforcing_anthro), 'k', marker = '.', markeredgecolor = 'r', label = "RCP 8.5")
ax1.set_title("Anthropogenic radiative forcing")
ax1.set_ylabel("Anthropogenic forcing")
ax1.set_xlabel("Year")
ax1.legend()
ax1.xaxis.set_ticks([0, 30, 80, 135, 230])
ax1.xaxis.set_ticklabels([1870, 1900, 1950, 2005, 2100])
ax2.plot(np.array(df_rcp45_radforcing_anthro), df_rcp45_radforcing_anthro.index, 'b.', label = "RCP 4.5")
ax2.plot(np.array(df_rcp85_radforcing_anthro), df_rcp85_radforcing_anthro.index, 'r.', label = "RCP 4.5")
ax2.set_xlabel("Anthropogenic forcing")
ax2.set_ylabel("Year")
ax2.legend();
```

```python
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(18, 4), gridspec_kw = {'width_ratios':[3, 1]})
ax1.plot(np.array(df_rcp45_radforcing_all), 'k', marker = '.', markeredgecolor = 'b', label = "RCP 4.5")
ax1.plot(np.array(df_rcp85_radforcing_all), 'k', marker = '.', markeredgecolor = 'r', label = "RCP 8.5")
ax1.set_title("Total radiative forcing")
ax1.set_ylabel("Total forcing")
ax1.set_xlabel("Year")
ax1.legend()
ax1.xaxis.set_ticks([0, 30, 80, 135, 230])
ax1.xaxis.set_ticklabels([1870, 1900, 1950, 2005, 2100])
ax2.plot(np.array(df_rcp45_radforcing_all), df_rcp45_radforcing_all.index, 'b.', label = "RCP 4.5")
ax2.plot(np.array(df_rcp85_radforcing_all), df_rcp85_radforcing_all.index, 'r.', label = "RCP 4.5")
ax2.set_xlabel("Total forcing")
ax2.set_ylabel("Year")
ax2.legend();
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1,3, figsize=(18,4))
ax1.plot(np.array(df_rcp45_radforcing_volcanic), 'k', marker = '.', markeredgecolor = 'b')
ax1.set_title("Volcanic radiative forcing")
ax1.set_xlabel("Year")
ax1.xaxis.set_ticks([0, 30, 80, 135, 230])
ax1.xaxis.set_ticklabels([1870, 1900, 1950, 2005, 2100])

ax2.plot(np.array(df_rcp45_radforcing_solar), 'k', marker = '.', markeredgecolor = 'b')
ax2.set_title("Solar radiative forcing")
ax2.set_xlabel("Year")
ax2.xaxis.set_ticks([0, 30, 80, 135, 230])
ax2.xaxis.set_ticklabels([1870, 1900, 1950, 2005, 2100])

ax3.plot(np.array(df_rcp85_radforcing_anthro), 'k', marker = '.', markeredgecolor = 'b')
ax3.set_title("Anthropogenic radiative forcing")
ax3.set_xlabel("Year")
ax1.xaxis.set_ticks([0, 30, 80, 135, 230])
ax1.xaxis.set_ticklabels([1870, 1900, 1950, 2005, 2100]);
```
