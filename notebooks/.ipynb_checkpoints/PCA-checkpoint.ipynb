```python
def sin_solar(x, radForcing_df_rcp85_solar, freq):
#     f = -0.084 * np.sin(2 * pi * (x + 3.5)/freq)
    f = np.sin(2 * pi * (x) / freq)
    f = f / (np.max(f) - np.min(f)) * (np.max(radForcing_df_rcp85_solar) - np.min(radForcing_df_rcp85_solar)) + \
                 np.mean(radForcing_df_rcp85_solar)
    return f

def cos_solar(x, radForcing_df_rcp85_solar, freq):
#     f = -0.084 * np.cos(2 * pi * (x + 3.5)/freq)
    f = np.cos(2 * pi * (x) / freq)
    f = f / (np.max(f) - np.min(f)) * (np.max(radForcing_df_rcp85_solar) - np.min(radForcing_df_rcp85_solar)) + \
                 np.mean(radForcing_df_rcp85_solar)
    return f

freq = 11

fig = plt.figure(figsize = (17,2))
x = np.linspace(0, radForcing_df_rcp85_solar.shape[0], radForcing_df_rcp85_solar.shape[0])
plt.plot(x, radForcing_df_rcp85_solar, '.-', label = 'solar')

plt.plot(x, sin_solar(x, radForcing_df_rcp85_solar, freq), '.-', label = 'sin')
plt.plot(x, cos_solar(x, radForcing_df_rcp85_solar, freq), '.-', label = 'cos')
plt.xticks(np.arange(0, x.shape[0], 11))
plt.legend();
#plt.plot(x, np.sin(x/11))
```
